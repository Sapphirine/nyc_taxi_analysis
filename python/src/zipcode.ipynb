```python
import numpy as np
import pandas as pd
jan=pd.read_csv('../output/jan_data.csv', sep='\t', index_col=False)
```

```python
jan1 = jan.loc[map(lambda x,y: x >=(-75) and x<=(-72) and y >= 40 and y <= 41.5, \
                  jan.Longitude, jan.Latitude),:]
```

```python
print jan.shape
print jan1.shape
```

```python
jan1.index=range(0, jan1.shape[0])
```

```python
jan1.head()
```

```python
jan1["Latitude"] = jan1["Latitude"].astype(str)
jan1["Longitude"] = jan1["Longitude"].astype(str)
```

```python
jan1['Latitude1'] = map(lambda x: x[0:5], jan1['Latitude'])
```

```python
jan1['Longitude1'] = map(lambda x: x[0:6], jan1['Longitude'])
```

```python
jan1["Coordinate"] = jan1[["Latitude1", "Longitude1"]].apply(lambda x: ",".join(x), axis=1)
```

```python
print len(set(jan1["Coordinate"]))
print len(jan1["Coordinate"])
jan1.head()
```

```python
jan1.to_csv('../output/jan_zipcode_output.csv', sep='\t', encoding='utf-8', index_col=False)
```

```python
df=pd.DataFrame(list(set(jan1['Coordinate'])))
```

```python
from geopy.geocoders import Nominatim
import re
import time
import csv
import sys
geolocator = Nominatim()

for n in range(1868, df.shape[0]-1):
    coord = df[0][n]
    try:
        location = geolocator.reverse(coord, timeout=66)
        if location.address == None:
            zipcode = "NaN"
        else:
            num = str(re.findall(r"\D(\d{5})\D", location.address))
            num = str(filter(str.isdigit, num))
            zipcode = num
        output = str(coord) + "\t" + zipcode
        print str(n) + " " + output
    #     time.sleep(0.2)

        with open(r"../output/zipcode_output.csv", 'a') as f:
            writer = csv.writer(f, delimiter='\t', lineterminator='\n')
            writer.writerow([output])
    except:
        print 'exception: ', sys.exc_info()[0]
        print coord
```
